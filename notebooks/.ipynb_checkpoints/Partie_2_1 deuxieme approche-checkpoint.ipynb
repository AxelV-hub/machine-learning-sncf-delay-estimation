# Etape 1 : régression après un one-hot encoding sur les gares

L'objectif de la première étape est d'entrainer de premiers modèles de régression à partir d'un simple one-hot encoding des gares de départ et gares d'arrivée. Un preprocessing classique sera effectué sur les autres features. 

### Preprocessing 

On commence par faire une copie du data set brute, et on ne garde que les colonnes qui nous intéressent pour ce problème, à savoir les colonnes que l'on se donne comme entrée : 
- la date (le mois et l'année concernée)
- la gare de départ et la gare d'arrivée
- le service (national ou international)
- la nombre de trains prévus sur le mois
- la durée moyenne du trajet sur le mois 

et la colonne que l'on essaye de prédire : 
- retard moyen tous trains arrivee

```python
data1 = data_brute.copy()

#on ne garde que les colonnes qui seront utilisées pour cette partie
data1 = data1.filter(items = ['date', 'service', 'gare_depart', 'gare_arrivee', 'duree_moyenne','nb_train_prevu','retard_moyen_tous_trains_arrivee'])

#on transforme en type date la colonne date pour appliquer des masks pandas plus facilement, pratique pour le split train-test
data1['date_object'] = pd.to_datetime(data1['date']) 
```

Définition de fonctions utiles pour les preprocessing suivants

```python
#applique un one_hot_encoder sur la liste de colonnes donnée en deuxième argument, renvoie le nouveau dataset. 
#l'argument drop_one permet de choisir ou non de supprimer une des colonnes pour éviter tout problème de combinaison linéaire
#la fonction retourne également la liste des noms des colonnes ajoutées
def one_hot_encoder(df,columns,drop_one = False):
    if drop_one:
        liste_colonnes = []
        for col in columns:
            liste_colonnes+=[col+'_'+value for value in pd.unique(df[col])[1:]]
        return pd.concat([df]+[pd.get_dummies(df[col],prefix=col) for col in columns],axis=1).drop(columns+[col+'_'+df.loc[0,col] for col in columns],axis=1),liste_colonnes
    else :
        liste_colonnes = []
        for col in columns:
            liste_colonnes+=[col+'_'+value for value in pd.unique(df[col])]
        return pd.concat([df]+[pd.get_dummies(df[col],prefix=col) for col in columns],axis=1).drop(columns,axis=1),liste_colonnes
```

```python
#définition d'une fonction qui supprime les lignes avec des outliers pour la colonne passée en deuxième argument, IQR_limit est le multiple de l'interquantile au delà duquel on considère que c'est un outlier
def remove_outliers(df,column,IQR_limit):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - IQR_limit * IQR
    upper_bound = Q3 + IQR_limit * IQR

    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
#définition d'une fonction qui applique un scaler_type sur la liste de colonnes passée en 4e argument
#le set d'entrainement est utilisé pour fit le scaler, puis le scaler est utilisé sur le set d'entrainement et le set de test
def scaler_applier(train_set, test_set, scaler_type, columns):
    train_set_copy = train_set.copy(deep=True)  
    test_set_copy = test_set.copy(deep=True)  
    for col in columns:
        scaler = scaler_type()
        data_train = np.array(train_set_copy[col]).reshape(-1, 1)
        data_test = np.array(test_set_copy[col]).reshape(-1, 1)
        scaler.fit(data_train)
        train_set_copy.loc[:, col] = scaler.transform(data_train)
        test_set_copy.loc[:, col] = scaler.transform(data_test)
    return train_set_copy, test_set_copy

#une autre version qui applique un scaler_type sur un data_set complet 
def scaler_applier_fulldata(data, scaler_type, columns):
    data_copy = data.copy(deep=True)  
    for col in columns:
        scaler = scaler_type()
        data_col = np.array(data_copy[col]).reshape(-1, 1)
        scaler.fit(data_col)
        data_copy.loc[:, col] = scaler.transform(data_col)
    return data_copy
```

```python
#définition d'une fonction qui applique un encoder simple sur la liste de colonnes passée en 4e argument
#le set d'entrainement est utilisé pour fit l'encoder, puis l'encoder est utilisé sur le set d'entrainement et le set de test
def encoder_applier(train_set, test_set, encoder_type, columns):
    train_set_copy = train_set.copy(deep=True)  
    test_set_copy = test_set.copy(deep=True)   
    for col in columns:
        encoder = encoder_type()
        encoder.fit(train_set_copy[col])
        train_set_copy.loc[:, col] = encoder.transform(train_set_copy[col])
        test_set_copy.loc[:, col] = encoder.transform(test_set_copy[col])
    return train_set_copy, test_set_copy

#une autre version qui applique un encoder_type sur un data_set complet 
def encoder_applier_fulldata(data, encoder_type, columns):
    data_copy = data.copy(deep=True)  
    for col in columns:
        encoder = encoder_type()
        encoder.fit(data_copy[col])
        data_copy.loc[:, col] = encoder.transform(data_copy[col])
    return data_copy  
```

On applique les preprocessing suivants : 
- Les colonnes `gare_depart` et `gare_arrivee` sont des données catégorielles à plus de 2 classes, on utilise donc un one-hot encoding sur ces deux colonnes afin d'obtenir des dummy colonnes. On prend soin de supprimer une dummy colonne pour éviter le problème de combinaisons linéaires pouvant causer des erreurs pour la régression linéaire. On fait cette étape avant le split entre train_set et test_set afin d'avoir la même liste de gares des deux cotés, et la même colonne supprimée
- On encode la `date` par encoding (et non en one-hot pour le moment). La date est donc vu comme une variable linéaire mais non cyclique. Puis on appliquer un scaler MinMax (avant la séparation du `train_set` et `test_set` afin d'éviter des dates du test set non connues par le train set)
- On separe le `train_set` du `test_set` en fonction de la date : avant et après janvier 2023
- On supprime les instances avec des valeurs extrêmes de `retard_moyen_tous_trains_arrivee` afin de supprimer les événements trop exceptionnels dans notre entrainement, ici avec un écart avec la médiane plus grand que 2x l'interquantile
- La feature `service` ne peut prendre que deux valeurs qui seront codés en 0 et 1
- Enfin, les `duree_moyenne`, `nb_train_prevu` et les `retard_moyens` sont normalisés avec la méthode MinMax

```python
data1.head()
```

```python
#on commence par one-hot encoder le data set complet pour éviter les problèmes de gares dans le test_set mais pas dans le train_set
data1,liste_gares = one_hot_encoder(data1,['gare_depart','gare_arrivee'],drop_one = True)

#on encode la colonnes de date avant le split train_set et test_set pour lesquels les dates seront disjointes
data1 = encoder_applier_fulldata(data1,LabelEncoder,['date'])
data1 = scaler_applier_fulldata(data1,MinMaxScaler,['date'])

#on sépare en train_set et test_set en fonction de la date avant et après 2023
data_train = data1[ data1['date_object']<= '2022-12' ]
data_test = data1[ data1['date_object']>= '2023-01' ]

#on supprime les outliers pour le train_set 
data_train = remove_outliers(data_train,'retard_moyen_tous_trains_arrivee',2)

#on normalise le train_set et le test_set 
data_train,data_test = encoder_applier(data_train,data_test,LabelEncoder,['service'])
data_train,data_test = scaler_applier(data_train,data_test,MinMaxScaler,['duree_moyenne','nb_train_prevu'])
```

```python
#on sépare les colonnes connues de la colonne à estimer
X_train = data_train.filter(items = ['date', 'service', 'duree_moyenne','nb_train_prevu']+liste_gares,axis = 1)
X_test = data_test.filter(items = ['date', 'service', 'duree_moyenne','nb_train_prevu']+liste_gares,axis = 1)
y_train = data_train['retard_moyen_tous_trains_arrivee']
y_test = data_test['retard_moyen_tous_trains_arrivee']
```

```python
X_train.head()
```

### Application des modèles de régression 

On utilisera 3 modèles de régression : la régression linéaire, le random forest, et le KNN. Ces 3 modèles seront également repris dans les parties suivantes afin d'être comparés. 

Les hyperparamètres des différents modèles de régression ont été estimés en utilisant une méthode RandomizedSearchCV dont le code sera ci-dessous diesé `#`. 

On commence par définir une fonction qui permettra d'estimer l'erreur quadratique de chaque model

```python
def evaluate(model, test_features, test_labels):
    predictions = model.predict(test_features).reshape(len(test_labels), 1)
    rmse = mean_squared_error(test_labels, predictions, squared=False)
    return rmse
```

# Etape 2 : régression avec une représentation dans un espace de dimension plus faible

Dans cette partie, nous allons tenter de créer du sens sur les données des gares, en "rapprochant" les gares pour lesquelles il y a des retards en même temps. Cela pourrait s'interpréter dans la réalité par des gares sur une même ligne, ou dans une même région géographique qui subiraient les mêmes retards liés à des traveaux, intempéries, accidents etc. 

Pour faire ceci, on crée un nouveau data set `data_reduc_dim` spécifique sur lequel on pourra appliquer une réduction de dimension. Chaque ligne représentera un mois, chaque colonne représentera une gare de départ ou une gare d'arrivée, et le data set condiendra le retard moyen sur le mois pour cette gare de départ ou d'arrivée.

N'ayant que 60 mois de data à disposition pour ~120 gares, le problème sera séparé en 2 : les gares de départ d'un côté, et les gares d'arrivée de l'autre. 

L'algorithme du PCA sera utilisé pour la réduction de dimension. Grâce aux composantes principales obtenues, nous pourrons ensuite projeter les gares de la partie 1 sur un espace de dimension plus faible et appliquer les 3 models de régression.  

### Preprocessing pour la réduction de dimension

```python
data_reduc_dim_gare_depart = data_brute.copy()
data_reduc_dim_gare_arrivee = data_brute.copy()

#on transforme en type date la colonne date pour appliquer des masks pandas plus facilement, pratique pour le split train-test
data_reduc_dim_gare_depart['date_object'] = pd.to_datetime(data_reduc_dim_gare_depart['date']) 
data_reduc_dim_gare_arrivee['date_object'] = pd.to_datetime(data_reduc_dim_gare_arrivee['date']) 

#on ne garde que les colonnes qui seront utilisées pour cette partie
data_reduc_dim_gare_depart = data_reduc_dim_gare_depart.filter(items = ['date_object', 'gare_depart', 'retard_moyen_tous_trains_arrivee'])
data_reduc_dim_gare_arrivee = data_reduc_dim_gare_arrivee.filter(items = ['date_object', 'gare_arrivee', 'retard_moyen_tous_trains_arrivee'])

#pour la réduction de dimension, on ne travaille que sur les données d'entrainement pour ne pas biaiser les résultats
data_reduc_dim_gare_depart = data_reduc_dim_gare_depart[data_reduc_dim_gare_depart['date_object']<='2022-12']
data_reduc_dim_gare_arrivee = data_reduc_dim_gare_arrivee[data_reduc_dim_gare_arrivee['date_object']<='2022-12']

data_reduc_dim_gare_depart.head()
```

```python
#permet de créer un tableau avec les valeurs de value_column en fonction des paramètres donnés par groupby_column en ligne, et class_column en colonne
def group_by(data,class_column,value_column,groupby_column):
    data,liste_values = one_hot_encoder(data,[class_column],drop_one = False)
    for colonne in liste_values:
        data[colonne].apply(lambda x : None if (x==0) else x)
        data[colonne]=data[value_column]*data[colonne]  
    data = data.groupby([groupby_column]).mean()
    data = data.drop([value_column],axis = 1)
    data = data.reset_index()
    return data,liste_values
```

```python
#on utilise la fonction group_by pour calculer le retard moyen par gare et par mois
data_reduc_dim_gare_depart,gare_depart_liste = group_by(data_reduc_dim_gare_depart,'gare_depart','retard_moyen_tous_trains_arrivee','date_object')
data_reduc_dim_gare_arrivee,gare_arrivee_liste = group_by(data_reduc_dim_gare_arrivee,'gare_arrivee','retard_moyen_tous_trains_arrivee','date_object')

#plus besoin de l'information de la date, on réinitialise l'index
data_reduc_dim_gare_depart = data_reduc_dim_gare_depart.drop(['date_object'],axis=1)
data_reduc_dim_gare_arrivee = data_reduc_dim_gare_arrivee.drop(['date_object'],axis=1)

#on normalise les données pour qu'elles soient centrées réduites 
data_reduc_dim_gare_depart = scaler_applier_fulldata(data_reduc_dim_gare_depart,StandardScaler,gare_depart_liste)
data_reduc_dim_gare_arrivee = scaler_applier_fulldata(data_reduc_dim_gare_arrivee,StandardScaler,gare_arrivee_liste)

data_reduc_dim_gare_arrivee.head()
```

### Application du PCA sur le data set des gares de départ et le data set des gares d'arrivée

```python
#On applique le PCA aux deux data sets
pca_depart = PCA()
pca_arrivee = PCA()

pca_depart_result = pca_depart.fit_transform(data_reduc_dim_gare_depart)
pca_arrivee_result = pca_arrivee.fit_transform(data_reduc_dim_gare_arrivee)
```

```python
explained_variance_depart = pca_depart.explained_variance_ratio_

plt.plot(range(1, len(explained_variance_depart) + 1), explained_variance_depart, marker='o')
plt.xlabel('Component Number')
plt.ylabel('Explained Variance')
plt.title('[Départs] Explained Variance by Component Number')
plt.show()
```

```python
top_10_explained = sum(explained_variance_depart[0:10])
print("Les 10 premières composantes principales permettent d'expliquer "+str(int(top_10_explained * 100))+"% de la variance")
```

```python
explained_variance_arrivee = pca_arrivee.explained_variance_ratio_

plt.plot(range(1, len(explained_variance_arrivee) + 1), explained_variance_arrivee, marker='o')
plt.xlabel('Component Number')
plt.ylabel('Explained Variance')
plt.title('[Arrivée] Explained Variance by Component Number')
plt.show()
```

```python
top_10_explained = sum(explained_variance_arrivee[0:10])
print("Les 10 premières composantes principales permettent d'expliquer "+str(int(top_10_explained * 100))+"% de la variance")
```

Les courbe de variances expliquées montrent que les premières composantes principales ont un poids important. On ne gardera que les 10 premières composantes principales afin de faire les projections de la partie suivante.

### Preprocessing et projection sur le nouvel espace à dimension réduite

On applique d'abord le même type de preprocessing qu'à l'étape précédente, en ajoutant la projection des vecteurs obtenus par le one-hot-encoding des gares sur les 10 composantes principales du PCA.

```python
data2 = data_brute.copy()

#on ne garde que les colonnes qui seront utilisées pour cette partie
data2 = data2.filter(items = ['date', 'service', 'gare_depart', 'gare_arrivee', 'duree_moyenne','nb_train_prevu','retard_moyen_tous_trains_arrivee'])

#on transforme en type date la colonne date pour appliquer des masks pandas plus facilement, pratique pour le split train-test
data2['date_object'] = pd.to_datetime(data2['date']) 


data2.head()
```

```python
#on commence par one-hot encoder le data set complet pour éviter les problèmes de gares dans le test_set mais pas dans le train_set
data2,liste_gares = one_hot_encoder(data2,['gare_depart','gare_arrivee'],drop_one = False)

#on encode la colonnes de date avant le split train_set et test_set pour lesquels les dates seront disjointes
data2 = encoder_applier_fulldata(data2,LabelEncoder,['date'])
data2 = scaler_applier_fulldata(data2,MinMaxScaler,['date'])

#on sépare en train_set et test_set en fonction de la date avant et après 2023
data_train = data2[ data2['date_object']<= '2022-12' ]
data_test = data2[ data2['date_object']>= '2023-01' ]

#on supprime les outliers pour le train_set 
data_train = remove_outliers(data_train,'retard_moyen_tous_trains_arrivee',2)

#on normalise le train_set et le test_set 
data_train,data_test = encoder_applier(data_train,data_test,LabelEncoder,['service'])
data_train,data_test = scaler_applier(data_train,data_test,MinMaxScaler,['duree_moyenne','nb_train_prevu'])

#on sépare les colonnes connues de la colonne a estimer
X_train = data_train.filter(items = ['date', 'service', 'duree_moyenne','nb_train_prevu']+liste_gares,axis = 1)
X_test = data_test.filter(items = ['date', 'service', 'duree_moyenne','nb_train_prevu']+liste_gares,axis = 1)
y_train = data_train['retard_moyen_tous_trains_arrivee']
y_test = data_test['retard_moyen_tous_trains_arrivee']

X_train.head()
```

```python
#projection des données des gares sur les composantes principales
projected_depart_train = np.dot(X_train.filter(items = gare_depart_liste), pca_depart.components_[:10].T)
projected_arrivee_train = np.dot(X_train.filter(items = gare_arrivee_liste), pca_arrivee.components_[:10].T)
projected_depart_test = np.dot(X_test.filter(items = gare_depart_liste), pca_depart.components_[:10].T)
projected_arrivee_test = np.dot(X_test.filter(items = gare_arrivee_liste), pca_arrivee.components_[:10].T)

#ajout des projections dans le data_set, appelé maintenant avec le suffix short car à plus faible dimension
X_train_short = pd.concat([X_train.reset_index(drop=True), pd.DataFrame(data=projected_depart_train, columns=[f'PC{i+1}_depart' for i in range(10)]),pd.DataFrame(data=projected_arrivee_train, columns=[f'PC{i+1}_arrivee' for i in range(10)])],axis = 1)
X_train_short.drop(gare_depart_liste+gare_arrivee_liste,axis = 1,inplace = True)
X_test_short = pd.concat([X_test.reset_index(drop=True), pd.DataFrame(data=projected_depart_test, columns=[f'PC{i+1}_depart' for i in range(10)]),pd.DataFrame(data=projected_arrivee_test, columns=[f'PC{i+1}_arrivee' for i in range(10)])],axis = 1)
X_test_short.drop(gare_depart_liste+gare_arrivee_liste,axis = 1,inplace = True)

X_train_short.head()
```

### Application des modèles de régression

Comme pour la partie 1, nous utilisons les 3 modèles de régression avec les nouvelles données obtenues par la réduction de dimension.

**Linear Regression**

# Etape 3 : régression avec une représentation dans un espace de dimension plus faible et du feature engineering sur les dates

Dans cette partie, nous utiliserons la réduction de dimension de l'étape 2 ainsi que les méthodes de feature engineering sur les dates de l'approche 1 (voir Notebook approche 1)

### Preprocessing des dates

```python
data3 = data_brute.copy()

#on ne garde que les colonnes qui seront utilisées pour cette partie
data3 = data3.filter(items = ['date', 'service', 'gare_depart', 'gare_arrivee', 'duree_moyenne','nb_train_prevu','retard_moyen_tous_trains_arrivee'])

#on transforme en type date la colonne date pour appliquer des masks pandas plus facilement, pratique pour le split train-test
data3['date_object'] = pd.to_datetime(data3['date']) 

#on divise la date en deux colonnes : année et mois
data3['annee'] = data3['date_object'].dt.year
data3['mois'] = data3['date_object'].dt.month

data3.head()
```

Pour le feature engineering sur les dates, on crée deux fonctions permettant d'ajouter les retards des `n_months=3` derniers mois comme nouvelles features pour chaque instance, et le retard moyen de tous les mois passés pour chaque ligne. 

```python
#définition d'une fonction qui ajoute pour chaque instance, les retards moyens des n_months derniers mois sur la même ligne de train
def add_delays_previous_months(df, n_months):
    print("Adding delays of previous months, this can take a while")
    n_instances = df.shape[0]
    past_month_cols = [f"retards_moys_mois_n-{i}" for i in range(1, n_months+1)]
    for col in past_month_cols:
        df[col] = 0.0
    for i,row in df.iterrows():
        if int(100*(i+1)/n_instances) % 10 == 0 and not(int(100*i/n_instances) % 10 == 0):
            print(str(int(100*(i+1)/n_instances))+"% completed")
        annee, mois = row["annee"], row["mois"]
        gare_depart, gare_arrivee = row["gare_depart"], row["gare_arrivee"]
        for j, col_name in enumerate(past_month_cols):
            n_mois_a_soustraire = j+1
            date_mois_prec = (annee, mois-n_mois_a_soustraire) if mois>=n_mois_a_soustraire+1 else ((annee-1, 12+mois-n_mois_a_soustraire) if annee>=2019 else (annee, mois))
            # print((annee, mois), n_mois_a_soustraire, date_mois_prec)
            series = df.loc[(df.gare_depart == gare_depart) & (df.gare_arrivee == gare_arrivee) & (df.annee == date_mois_prec[0]) & (df.mois == date_mois_prec[1]), "retard_moyen_tous_trains_arrivee"]
            # print(series)
            if not series.empty:
                value = series.values[0]
                # print(value)
                df.loc[i, col_name] = value
            else:
                # un peu brutal de prendre la moyenne totale direct : on pourrait essayer le mois d'avant encore ou l'année d'avant...
                df.loc[i, col_name] = row["retards_moy_arrivee_trajet_passés"]
    return df
```

```python
#définition d'une fonction qui ajouter pour chaque instance, la moyenne des retards pour tous les trajets passés sur la même ligne (sur les données disponibles)
def add_past_avg_delay(df):

    df["retards_moy_arrivee_trajet_passés"] = 0.0

    for i, row in df.iterrows():
        gare_depart, gare_arrivee = row["gare_depart"], row["gare_arrivee"]
        if i>1:
            mean_delay_series = df[:i-1].loc[(df.gare_depart == gare_depart) & (df.gare_arrivee == gare_arrivee),"retard_moyen_tous_trains_arrivee"]
            # print(mean_delay)
            if mean_delay_series.empty:
                # on prend la même valeur pour la première occurence de chaque trajet
                df.loc[i, "retards_moy_arrivee_trajet_passés"] = row["retard_moyen_tous_trains_arrivee"]
            else:
                mean_delay =  mean_delay_series.mean()
                df.loc[i, "retards_moy_arrivee_trajet_passés"] = mean_delay
        else:
            # on prend la même valeur pour la première occurence de chaque trajet (qui est aussi la première ligne du df dans ce cas)
            df.loc[i, "retards_moy_arrivee_trajet_passés"] = row["retard_moyen_tous_trains_arrivee"]

    return df
```

```python
#on ajoute le retard moyen du passé pour chaque instance
data3 = add_past_avg_delay(data3)

#on ajoute le retard moyen des 3 mois précédents pour chaque trajet 
n_months = 3
data3 = add_delays_previous_months(data3, n_months)

data3.head()
```

On utilise cette fois ci un one-hot encoder sur les mois de l'année

```python
#on transforme les mois pour correspondre directement à leur nom en string
month_mapping = {
    1: 'Janvier',
    2: 'Février',
    3: 'Mars',
    4: 'Avril',
    5: 'Mai',
    6: 'Juin',
    7: 'Juillet',
    8: 'Août',
    9: 'Septembre',
    10: 'Octobre',
    11: 'Novembre',
    12: 'Décembre'
}
data3['mois'] = data3['mois'].map(month_mapping)

#on commence par one-hot encoder le data set complet pour éviter les problèmes de gares dans le test_set mais pas dans le train_set
data3,liste_gares = one_hot_encoder(data3,['gare_depart','gare_arrivee'],drop_one = False)
data3,liste_mois = one_hot_encoder(data3,['mois'],drop_one = True)

#on scale les colonnes mois et années le split train_set et test_set pour lesquels les dates seront disjointes
data3 = scaler_applier_fulldata(data3,MinMaxScaler,['annee'])
```

### Preprocessing des autres features

```python
#on sépare en train_set et test_set en fonction de la date avant et après 2023
data_train = data3[ data3['date_object']<= '2022-12' ]
data_test = data3[ data3['date_object']>= '2023-01' ]

#on supprime les outliers pour le train_set 
data_train = remove_outliers(data_train,'retard_moyen_tous_trains_arrivee',2)

#on normalise le train_set et le test_set 
data_train,data_test = encoder_applier(data_train,data_test,LabelEncoder,['service'])
data_train,data_test = scaler_applier(data_train,data_test,MinMaxScaler,['duree_moyenne','nb_train_prevu','retards_moys_mois_n-1','retards_moys_mois_n-2','retards_moys_mois_n-3','retards_moy_arrivee_trajet_passés'])

#on sépare les colonnes connues de la colonne a estimer
X_train = data_train.filter(items = ['annee', 'service', 'duree_moyenne','nb_train_prevu','retards_moys_mois_n-1','retards_moys_mois_n-2','retards_moys_mois_n-3','retards_moy_arrivee_trajet_passés']+liste_gares+liste_mois,axis = 1)
X_test = data_test.filter(items = ['annee', 'service', 'duree_moyenne','nb_train_prevu','retards_moys_mois_n-1','retards_moys_mois_n-2','retards_moys_mois_n-3','retards_moy_arrivee_trajet_passés']+liste_gares+liste_mois,axis = 1)
y_train = data_train['retard_moyen_tous_trains_arrivee']
y_test = data_test['retard_moyen_tous_trains_arrivee']
```
